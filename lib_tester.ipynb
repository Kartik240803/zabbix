## Get Alerts With Ack

```python
result = zbx.get_alerts(
    hostname=hostname,
    time_from=time_from,
    time_to=time_to
)
result  
print("\nOutput is Converted to dataframe")
d = pd.DataFrame(result)
event_ids = d['eventid'].dropna().unique().tolist()
ack_dict = zbx.get_acknowledgement(event_ids).get('data')  # returns {eventid: data}
ack_df = pd.DataFrame(ack_dict)

if ack_dict:
    ack_df = pd.DataFrame(ack_dict)
    ack_df['clock'] = pd.to_numeric(ack_df['clock'], errors='coerce').astype('Int64')
    ack_df = ack_df.rename(columns={
        'clock': 'acknowledged_time',
        'message': 'acknowledged_message',
        'userid': 'acknowledged_by',
        'name': 'acknowledged_by_name'  # optional rename for clarity
    })
    d = d.merge(ack_df, on='eventid', how='left')
else:
    # If ack_dict is empty, add blank columns manually
    d['acknowledged_time'] = pd.NA
    d['acknowledged_message'] = pd.NA
    d['acknowledged_by'] = pd.NA
    d['acknowledged_by_name'] = pd.NA  # only if 'name' is expected
d
```

## Get First time to response and resolution time

```python
# First Time to Response = acknowledged_time - start_time
d['first_time_to_response'] = d['acknowledged_time'] - d['start_time']

# Time to Resolution = end_time - start_time only if recovery_eventid is not NaN
d['time_to_resolution'] = d.apply(
    lambda row: row['end_time'] - row['start_time'] if pd.notna(row['recovery_eventid']) else None,
    axis=1
)
d.head(2)
```

### Get grouped ['Hourly','Daily','Weekly','Montly'] data for First time to response and resolution time 


```python
def group_response_resolution(df, interval):
    # Convert UNIX time to datetime only for grouping
    df['start_dt'] = pd.to_datetime(df['start_time'], unit='s')

    # Compute response and resolution times (in seconds)
    df['first_time_to_response'] = df['acknowledged_time'] - df['start_time']
    df['time_to_resolution'] = df.apply(
        lambda row: row['end_time'] - row['start_time'] if pd.notna(row['recovery_eventid']) else None,
        axis=1
    )

    # Define how to group based on interval
    if interval == 'hourly':
        df['group_time'] = df['start_dt'].dt.floor('h')
    elif interval == 'daily':
        df['group_time'] = df['start_dt'].dt.date
    elif interval == 'weekly':
        df['group_time'] = df['start_dt'].dt.to_period('W').apply(lambda r: r.start_time)
    elif interval == 'monthly':
        df['group_time'] = df['start_dt'].dt.to_period('M').apply(lambda r: r.start_time)
    else:
        raise ValueError("Invalid interval. Choose from: 'hourly', 'daily', 'weekly', 'monthly'.")

    # Group and aggregate
    result = df.groupby('group_time').agg({
        'start_time': 'count',
        'first_time_to_response': 'mean',
        'time_to_resolution': 'mean'
    }).reset_index().rename(columns={'start_time': 'count'})
    result = result.fillna(0)

    return result
```

```python
interval = "daily"  # Change this to "hourly", "daily", "monthly" as needed
result_df = group_response_resolution(d, interval)
print(result_df)
```

## Get Similar Hosts from host name 

```python
result = zbx.get_host_by_group(host_group='all')
result
d = pd.DataFrame(result)
def get_similar_hosts(row, df):
    same_group_hosts = df[df["group_name"] == row["group_name"]]["host_name"].tolist()
    return [host for host in same_group_hosts if host != row["host_name"]]

# Apply function to each row
d["similar_hosts"] = d.apply(lambda row: get_similar_hosts(row, d), axis=1)

# Optional: make sure full output is shown in console
pd.set_option('display.max_colwidth', None)

d
```
