```python
import ZabFetch as zb
import time
import pandas as pd
import os
import json
import filecmp
```

```python
config_path = os.path.join('/root/zabbix_db_lib/', 'config.json')
with open(config_path, 'r') as f:
    config = json.load(f)
    db_config = config['database_config']
```

```python
zbx= zb.ZabbixDB(**db_config)
```

##  Setting Test variables

```python
hostname = 'Zabbix server'
metric_name = 'CPU utilization'
time_to = int(time.time())
time_from = time_to - 86400
itemid = 42269
```

## Testing Get Item Details Function

### With HostName

```python

result = zbx.get_item_detail(
    hostname=hostname,
    item_name=metric_name
)
print(result)
```

### Without HostName

```python

result = zbx.get_item_detail(
    item_name=metric_name
)
print(result)
```

## Get Data

```python

result = zbx.get_history_data(
    itemid=itemid,
    time_from=time_from,
    time_to=time_to,
    history_table_name='history'
)
print(result)

print("\nOutput is Converted to dataframe")
d = pd.DataFrame(result)
d
```

```python

result = zbx.get_trend_data(
    itemid=itemid,
    time_from=time_from,
    time_to=time_to,
    trend_table_name='trends',
    statistical_measure='all'

)
print(result)

print("\nOutput is Converted to dataframe")
d = pd.DataFrame(result)
d
```

This is the valid statistical_measure ['min', 'max', 'mean', 'median', 'stdev', 'sum', 'count', 'range', 'mad', 'last', 'avg']

```python

result = zbx.get_metric_data(
    hostname=hostname,
    metric_name=metric_name,
    time_from=time_from,
    time_to=time_to
)
print(result)

print("\nOutput is Converted to dataframe")
d = pd.DataFrame(result['data'])
d
```
